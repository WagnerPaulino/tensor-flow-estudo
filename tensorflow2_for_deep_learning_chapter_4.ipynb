```
import deepchem as dc
import numpy as np
import pandas as pd
_, (train, valid, test), _ = dc.molnet.load_tox21()
train_X, train_y, train_w = train.X, train.y, train.w
valid_X, valid_y, valid_w = valid.X, valid.y, valid.w
test_X, test_y, test_w = test.X, test.y, test.w

# Remove extra tasks
train_X = np.array(train_X[:, 0], dtype=np.float)
train_y = np.array(train_y[:, 0], dtype=np.float)
valid_y = np.array(valid_y[:, 0], dtype=np.float)
test_y = np.array(test_y[:, 0], dtype=np.float)
train_w = np.array(train_w[:, 0], dtype=np.float)
valid_w = np.array(valid_w[:, 0], dtype=np.float)
test_w = np.array(test_w[:, 0], dtype=np.float)

```

```
%load_ext tensorboard

%tensorflow_version 2.x
import tensorflow as tf
import matplotlib.pyplot as plt

model = tf.keras.Sequential([
  tf.keras.layers.Dense(1,activation=tf.keras.activations.sigmoid)
])
```

```
model.compile(optimizer=tf.keras.optimizers.SGD(),
              loss=tf.keras.losses.BinaryCrossentropy(),
              metrics=['accuracy'])
history = model.fit(train_X, train_y, epochs=100)
```
